# **Project Summary -**


In today’s highly competitive streaming landscape, data-driven insights play a crucial role in shaping content strategies and sustaining user engagement. This project focuses on performing an Exploratory Data Analysis (EDA) of Amazon Prime Video’s content library to understand content distribution, diversity, trends over time, audience perception, and regional contribution. The dataset used for this analysis consists of two CSV files containing information about titles and credits available on Amazon Prime Video in the United States, including metadata such as release year, genres, age certifications, IMDb and TMDB ratings, popularity metrics, and contributor details.

The analysis began with data understanding and cleaning, where missing values were handled thoughtfully to avoid misleading interpretations. Multi-valued categorical columns such as genres and production countries were preprocessed using splitting and exploding techniques to enable accurate aggregation and analysis. Feature engineering, such as creating decade-based groupings, helped uncover long-term trends in content evolution.

Through a series of visualizations, the project explored multiple dimensions of the content library. Initial analysis revealed that Amazon Prime’s catalog is dominated by movies, with TV shows forming a smaller but significant portion of the platform. A time-based analysis showed a sharp increase in content production after the early 2000s, reflecting Amazon Prime’s aggressive expansion strategy in response to rising competition in the OTT market.

Content diversity analysis highlighted that genres such as Drama, Comedy, and Action dominate the platform. While this indicates strong audience demand for these genres, it also suggests a potential risk of genre saturation. Age certification analysis showed that mature-rated content (TV-MA and R) tends to attract higher popularity but also exhibits greater variability in IMDb ratings, implying polarized audience responses. In contrast, general-audience content demonstrated more consistent ratings, indicating broader but less intense engagement.

A comparative study of IMDb and TMDB metrics revealed a moderate positive correlation between the two rating platforms, suggesting general alignment in audience perception. However, popularity metrics showed weak correlation with ratings, indicating that high engagement does not always translate into high perceived quality. A dual-axis analysis comparing content volume and quality over time revealed that while the number of titles has increased significantly, average IMDb ratings have remained relatively stable, with a slight decline in recent decades. This suggests that rapid expansion has not drastically compromised quality but highlights the importance of maintaining balance between quantity and quality.

Regional analysis demonstrated that content production is heavily concentrated in a few countries, particularly the United States and India, with limited representation from other regions. This presents an opportunity for Amazon Prime to expand its international content portfolio to attract diverse global audiences.

Overall, this project provides actionable insights into Amazon Prime Video’s content strategy. The findings emphasize the importance of balancing content volume with quality, diversifying genres and regions, and leveraging both popularity and quality metrics for informed decision-making. These insights can help stakeholders such as content strategists, marketing teams, and investors optimize content investments, enhance user engagement, and support sustainable platform growth.

#### Chart - 10 : Content Maturity Analysis

````python
# Chart - Age Certification vs Average IMDb Score

age_stats = (titles.dropna(subset=['age_certification'])
             .groupby('age_certification')
             .agg(avg_rating=('imdb_score', 'mean'),
                  count=('title', 'count'))
             .reset_index()
             .sort_values(by='avg_rating', ascending=False))

plt.figure(figsize=(10,6))
sns.barplot(data=age_stats, x='age_certification', y='avg_rating')
plt.title('Content Maturity vs Quality')
plt.xlabel('Age Certification')
plt.ylabel('Average IMDb Score')
plt.xticks(rotation=45)
plt.show()
````
